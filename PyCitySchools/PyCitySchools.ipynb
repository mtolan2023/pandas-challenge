# PyCity Schools Analysis

> The dataset provides information about our district's 15 high schools, attended by 39,170 students. The data provided allowed for analysis of school type (district and charter), school size, math and reading scores (by grade and by school), and school funding (by school and per capita). The breakdown of each area can be found below.

> As a whole, the district is displaying an Overall Passing Percentage of 65.17% ("passing" meaning scores above 70%), with Average Reading Scores of 81.87% and Math Scores of 78.98%. When the data is broken down by school, Cabrera High School has the highest Overall Passing Percentage at 91.33% and Rodriguez High School has the lowest, at 52.98%.
    
> Examining performance by School Type shows a clear gap between District and Charter Schools. All of the top performing schools in the district are Charters and all of the bottom performing schools are District schools. On average, the Overall Passing Rate is 90.43% for Charter Schools and 53.67% for District Schools. When looking at the average Math and Reading Scores between School Types, there is less than 10% difference in the average scores for each subject but the Passing Rate for both is considerably higher at Charter Schools. The "% Passing Math" Rate is 27% higher at Charter Schools and the "% Passing Reading" is 16% higher at Charters. These numbers feed the difference in "% Overall Passing". Because of this, it is important to take into account both the average scores and the Overall Passing Rate, since the average scores show performance that is much closer between District and Charter Schools, with a the overall averages less than 15 points over the threshold for "passing" (>=70). All scores are near the line of "passing" vs. "not passing" and the fact that Charter scores average higher above that line does not conclude that the students are performing 16-27% better at Charter Schools. While it is important to consider both Average Scores and Passing Rates to get a total picture of performance, it is also clear that Charter Schools are outperforming District schools.

> With the exception of Wilson High School,  Charter High Schools Schools are smaller than District Schools and the per capita spending is lower. It would be difficult to draw any specific conclusions without more information on the schools but additional information about spending, age of school buildings, staffing levels, and student/neighborhood demographics could provide richer insights into the reasons for the performance gap between school types. 

> Across the district’s High Schools, Reading and Math scores showed little change across grades. Regardless of the overall school performance, grades stay consistent between 9th and 12th grade. From this, one can conclude that there isn’t a gap in instruction within one grade level and it would be useful to also see data from Middle Schools and Elementary Schools in order to attempt to pinpoint where instruction could be improved. This also provides some additional insight into questions of District vs. Charter performance. Consistent scores across grades mean that District and Charter High Schools are not having a significantly positive or negative impact in student performance, rather they are just maintaining the previous scores of students as they move up from Middle Schools. There is no evidence that student grades decrease as they progress through a District High School or that there is an improvement through the grades at Charter Schools. 
    
> The data related to school spending provides some interesting insights. First, there is no correlation between funding per student and school size or type. It seems as though there should be more consistency within a single district in terms of per capita budget. Some additional information regarding the derivation of Budget numbers and how school dollars are being spent per school would be useful. The budget numbers become more confusing when looking at per student spending versus test scores. There is actually a reverse correlation between dollars spend and Average Math/Reading scores and % Overall Passing per school. One might assume that the more resources are committed to each student, the higher that student’s chance of success would be. This data shows the opposite, where schools with the least spending per student have the highest scores and passing rates. Again, more budgetary information would help clarify this trend. Perhaps the large District High Schools are spending significantly on athletics, arts, or other extra-curricular activities, while Charters focus more on core learning and meeting performance metrics. Further budget information could support or disprove these theories.
    
> In the data provided, School Size also has an impact on overall scores. With the exception of Wilson High School (the only Charter High School within the Large (2000-5000) category), the largest High Schools have the lowest scores and Percentage of Overall Passing. More information on these large schools could help explain this performance gap, especially with Wilson High School as a contrast point for District versus Charter scores. It would be helpful to know more about the student to teacher ratio, teacher age, tenure, student support resources, etc. in order to gain more insight. 
    
> Overall, the dataset provides some clear baseline numbers for school performance, budget and size. It makes it clear that there is significant difference in student success between Charter and District High Schools but more information on Budget, earlier eduction, community and staff would be necessary to draw clear conclusions on how/where to act.

```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
# Calculate the total number of unique schools
school_count = school_data["school_name"].size
school_count
```

```python
# Calculate the total number of students
student_count = student_data["student_name"].size
student_count
```

```python
# Calculate the total budget
total_budget = school_data["budget"].sum()
total_budget
```

```python
# Calculate the average (mean) math score
average_math_score = student_data["math_score"].mean()
average_math_score
```

```python
# Calculate the average (mean) reading score
average_reading_score = student_data["reading_score"].mean()
average_reading_score
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
passing_math_count = school_data_complete[(school_data_complete["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(student_count) * 100
passing_math_percentage
```

```python
# Calculate the percentage of students who passed reading (hint: look at how the math percentage was calculated)  
passing_reading_count = school_data_complete[(school_data_complete["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage = passing_reading_count / float(student_count) * 100
passing_reading_percentage
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passing_math_reading_count = school_data_complete[
    (school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(student_count) * 100
overall_passing_rate
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
district_summary = pd.DataFrame({
    "Total Schools" : [school_count],
    "Total Students" : [student_count],
    "Total Budget" : [total_budget],
    "Average Math Score" : [average_math_score],
    "Average Reading Score" : [average_reading_score],
    "% Passing Math" : [passing_math_percentage],
    "% Passing Reading" : [passing_reading_percentage],
    "% Overall Passing" :[overall_passing_rate]
})



district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)

# Display the DataFrame
district_summary
```

### School Summary

School Types in District

```python
#Extract school name and type columns from main df and remove duplicates to obtain a list of school/type
school_types = school_data_complete.loc[:, ["school_name", "type"]]
school_types = school_types.drop_duplicates(keep= 'first')
school_types = school_types.sort_values('school_name')

school_types
```

Student Count Per School

```python
# Calculate the total student count per school
per_school_counts = school_data_complete.groupby(['school_name']).size()
per_school_counts
```

Per School Budget

```python
# Calculate the total school budget and per capita spending per school
# Group by school name and budget columns
per_school_budget = school_data_complete.groupby(['school_name', 'budget']).count()
#limit df to just the indexes
per_school_budget = per_school_budget[[]]

per_school_budget 
```

Per Capita Spending per School

```python
#Pull the name size and budget columns from the main df
per_capita_spending = school_data_complete.loc[:, ["school_name", "size", "budget"]]
#Drop Duplicate values to just retain the school name, budget and school size for each unique school
per_capita_spending = per_capita_spending.drop_duplicates(keep= 'first')
#Divide spending by size to get per capita spending
per_capita_spending["per capita spending"] = per_capita_spending["budget"] /per_capita_spending["size"]
#sort alphabetically to retain alphabetization
per_capita_spending = per_capita_spending.sort_values('school_name')
#set school name as index for df
per_capita_spending = per_capita_spending.set_index('school_name')

per_capita_spending
```

Average Math and Reading Scores per school

```python
# Calculate the average test scores per school
# Find the mean of both scores by grouping by school name and then finding the mean of the scores by group
per_school_math = school_data_complete.groupby(['school_name'])["math_score"].mean()
per_school_reading = school_data_complete.groupby(['school_name'])["reading_score"].mean()

per_school_math
per_school_reading

# Merging both subjects together to display in on df
average_math_reading = pd.merge(per_school_math, per_school_reading, on=["school_name"])
average_math_reading
```

```python
# # Calculate the number of students per school with math scores of 70 or higher
students_passing_math = school_data_complete[school_data_complete["math_score"] >= 70]
school_students_passing_math = students_passing_math.groupby(["school_name"]).size()


school_students_passing_math
```

```python
# Calculate the number of students per school with reading scores of 70 or higher
students_passing_reading = school_data_complete[school_data_complete["reading_score"] >= 70]
school_students_passing_reading = students_passing_reading.groupby(["school_name"]).size()

school_students_passing_reading
```

```python
# Use the provided code to calculate the number of students per school that passed both math and reading with scores of 70 or higher
students_passing_math_and_reading = school_data_complete[
    (school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)]

school_students_passing_math_and_reading = students_passing_math_and_reading.groupby(["school_name"]).size()

school_students_passing_math_and_reading
```

```python
# Use the provided code to calculate the passing rates
per_school_passing_math = school_students_passing_math / per_school_counts * 100
per_school_passing_reading = school_students_passing_reading / per_school_counts * 100
overall_passing_rate = school_students_passing_math_and_reading / per_school_counts * 100


per_school_passing_math
per_school_passing_reading
overall_passing_rate

#Converting the above series to df so that they can be merged into on df
per_school_passing_math = per_school_passing_math.to_frame().reset_index()
per_school_passing_reading = per_school_passing_reading.to_frame().reset_index()
overall_passing_rate = overall_passing_rate.to_frame().reset_index()

#Merging 3 dfs to display all results together
per_school_passing = pd.merge(per_school_passing_math, per_school_passing_reading, on=['school_name'])
per_school_passing = pd.merge(per_school_passing, overall_passing_rate, on=['school_name'])

# Renaming columns of merged dfs to accurately label columns
per_school_passing = per_school_passing.rename(columns= {'0_x':'% Passing Math', '0_y':'% Passing Reading', 0:'% Overall Passing'})


per_school_passing
```

```python
# # Create a DataFrame called `per_school_summary` with columns for the calculations above.
# Merging above dataframes on the common "school_name" index
per_school_summary = pd.merge(school_types, per_capita_spending, on=["school_name"])
per_school_summary = pd.merge(per_school_summary, average_math_reading, on=["school_name"])
per_school_summary = pd.merge(per_school_summary, per_school_passing, on=["school_name"])

#Set df index to "school_name"
per_school_summary = per_school_summary.set_index('school_name')

#rename columns 
per_school_summary = per_school_summary.rename(columns={'type':'School Type', 'size':'Total Students', \
                'budget':'Total School Budget', 'per capita spending':'Per Student Budget', 'math_score':'Average Math Score', \
                'reading_score':'Average Reading Score'})
#remove "school_name" index label
per_school_summary.index.name = None


# # Formatting
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)

# Display the DataFrame
per_school_summary
```

## Scores by School Spending

```python
# Establish the bins 
bins = [0, 585, 630, 645, 680]
labels = [" <$585 ", " $585-630 ", " $630-645 ", " $645-680 "]
```

```python
# Create a copy of the school summary since it has the "Per Student Budget" 
school_spending_df = per_school_summary.copy()
```

```python
# Binning School Spending
# First strip the '$' from the "Per Student Budget" column so it can be calculated.
school_spending_df["Per Student Budget"] = school_spending_df["Per Student Budget"].str.replace('$', '', regex=True)
#Then convert column type from object to float so that it can be caluculated
school_spending_df["Per Student Budget"] = school_spending_df["Per Student Budget"].astype(float)
#binning
school_spending_df['Spending Ranges (Per Student)'] = pd.cut(school_spending_df['Per Student Budget'], bins, labels=[" <$585 ", " $585-630 ", " $630-645 ", " $645-680 "])

#Add the '$' formatting back into the "Per Student Budget" column
school_spending_df["Per Student Budget"] = school_spending_df["Per Student Budget"].map("${:,.2f}".format)

school_spending_df
```

```python
#  Calculate averages for the desired columns. 
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Math Score"].mean()
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean()
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Math"].mean()
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Reading"].mean()
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Overall Passing"].mean()
```

```python
# Assemble into DataFrame
spending_summary = pd.DataFrame({
    "Average Math Score" : spending_math_scores,
    "Average Reading Score" : spending_reading_scores,
    "% Passing Math" : spending_passing_math,
    "% Passing Reading" : spending_passing_reading,
    "% Overall Passing" : overall_passing_spending
})

# Display results
spending_summary 
```

```python
# Calculate averages for the desired columns. 
size_math_scores = per_school_summary.groupby(["School Size"])["Average Math Score"].mean()
size_reading_scores = per_school_summary.groupby(["School Size"])["Average Reading Score"].mean()
size_passing_math = per_school_summary.groupby(["School Size"])["% Passing Math"].mean()
size_passing_reading = per_school_summary.groupby(["School Size"])["% Passing Reading"].mean()
size_overall_passing = per_school_summary.groupby(["School Size"])["% Overall Passing"].mean()
```

```python
# Create a DataFrame called `size_summary` that breaks down school performance based on school size (small, medium, or large).
# Use the scores above to create a new DataFrame called `size_summary`
size_summary = pd.DataFrame({
    "Average Math Score" : size_math_scores,
    "Average Reading Score" : size_reading_scores,
    "% Passing Math" : size_passing_math,
    "% Passing Reading" : size_passing_reading,
    "% Overall Passing" : size_overall_passing
})
    

# Display results
size_summary
```
